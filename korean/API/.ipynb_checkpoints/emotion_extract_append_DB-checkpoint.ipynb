```python
from lyrics_emotion_analysis_KR import load_vocab , load_model , hybrid_emotion_analysis , hybrid_emotion_export_persent
from lyrics_preprocessing import EnglishOrKorean , isKorean_percent , isRemoveData
import os
from tqdm import tqdm
import pandas as pd
import openpyxl
```

```python
tqdm.pandas()
```

```python
music_analysis_query = """SELECT DISTINCT rma.music_idx, rmm.title, rmm.lyrics
FROM RENEW_music_analytics AS rma
LEFT JOIN RENEW_meta_music AS rmm ON rma.music_idx = rmm.music_idx"""

RENEW_meta_music_query = """SELECT DISTINCT rmm.music_idx , rmm.title , rmm.lyrics
FROM RENEW_meta_music AS rmm
WHERE rmm.lyrics <> 'None'"""
```

```python
vocab_path = "./asset/vocab_9class_500.csv"
stopwords_path = "./asset/stopwords.txt"
# 모델이 저장되어있는 경로
checkpoint_path = os.path.join("./", 'Bert_save_model', 'best_model')
```

```python
vocab = load_vocab(vocab_path,stopwords_path)
```

```python
# 모델 객체 생성
sentiment_model = load_model(checkpoint_path)
```

### DB 불러오기

```python
import pymysql
#indj 데이터 베이스 config
RDS_config = {
    # db 접속 config 입력
}

# data base connect
indj_db = pymysql.connect(**RDS_config, cursorclass=pymysql.cursors.DictCursor)
```

```python
# 커서 객체 생성
cursor = indj_db.cursor()

# SQL 테이블을 pandas Dataframe 으로 불러오기
sql = RENEW_meta_music_query
cursor.execute(sql)
res = cursor.fetchall()
lyrics_df = pd.DataFrame(res)
```

```python
# 외국어 & 한국어 구분
# 한국어 비율이 10% 이상이면 한국어로 판단함.
lyrics_df['IsKorean'] = lyrics_df['lyrics'].progress_map(lambda x : isKorean_percent(x,0.1))
```

```python
# 한국어 필터링
korean_lyrics_df = lyrics_df[lyrics_df['IsKorean'] == 1]
```

```python
# inst , MR 제거 밒 필터링
korean_lyrics_df['isInst'] = korean_lyrics_df['title'].map(lambda x : isRemoveData(x))
korean_lyrics_df = korean_lyrics_df[korean_lyrics_df['isInst'] == 0]
```

### 미리 뽑혀진 데이터 제외

```python
# 커서 객체 생성
cursor = indj_db.cursor()

# SQL 테이블을 pandas Dataframe 으로 불러오기
sql = """SELECT * FROM `indj_service`.`RENEW_lyrics_emotion_percent`;"""
cursor.execute(sql)
res = cursor.fetchall()
analysis_df = pd.DataFrame(res)
```

```python
# 전체 데이터 - 삽입되어진 데이터 index (차집합)
unique_list = list(set(korean_lyrics_df['music_idx'].to_list()) -  set(analysis_df['music_idx'].to_list()))
```

```python
# music_idx 로 index를 바꿔주고 loc 으로 이미 추출해야만 하는 데이터 필터링(삽입되어진 데이터 제외)
korean_lyrics_df = korean_lyrics_df.set_index('music_idx')
korean_lyrics_df = korean_lyrics_df.loc[unique_list]
```

```python
korean_lyrics_df.reset_index(inplace=True)
```

### 추출 시작

```python
# 추후 DataFrame 변환을 위해 dict 형태로 추출시작
emotion_result_dict = {}
```

```python
for i in tqdm(range(len(korean_lyrics_df))[40000:90000]):
    low = korean_lyrics_df.iloc[i]
    emotion_result = hybrid_emotion_export_persent(low['lyrics'],sentiment_model,vocab)
    try:
        # musict_idx 가 key 값 , emotion_result 가 value 값
        emotion_result_dict[low['music_idx']] = emotion_result
    except IndexError:
        print("{}번째 idx 에러 - 가사 : {}".format(i,low['lyrics']))
    
```

```python
emotion_result_df = pd.DataFrame(emotion_result_dict).T
```

```python
emotion_result_df
```

```python
# NaN 값 -> 0 
emotion_result_df = emotion_result_df.fillna(0)
```

```python
# 추출한 데이터 임시 저장
emotion_result_df.to_csv('./ko_emotion_data/ko_210702.csv')
```

### DB에 데이터 삽입

```python
# 커서 객체 생성
cursor = indj_db.cursor()

# SQL 테이블을 pandas Dataframe 으로 불러오기
sql = """SELECT * FROM `indj_service`.`RENEW_lyrics_emotion_percent`;"""
cursor.execute(sql)
res = cursor.fetchall()
percent_df = pd.DataFrame(res)
```

#### 중복체크

```python
# 이미 삽입되었던 값과 중복이 되어있는지 체크함
# lyrics_emotion_percent table은  primary key 로 music_idx가 설정되어있기 때문에 중복값이 있으면 에러 발생
unique_list = list(set(emotion_result_df.index.to_list()) - set(percent_df.music_idx.to_list()))
```

```python
emotion_result_df = emotion_result_df.loc[unique_list]
```

```python
emotion_result_df
```

```python
indj_db = pymysql.connect(**RDS_config, cursorclass=pymysql.cursors.DictCursor)
cursor = indj_db.cursor()
top_emotion_data = []
emotion_percent_data = []

for i in tqdm(range(len(emotion_result_df))):
    low = emotion_result_df.iloc[i]
    top_emotion_values = [
        low['emotion1'],
        low['emotion2'],
        low['emotion3'],
        low.name # low index
    ]
    emotion_percent_values = [
        low.name,
        low['5359'], # 사랑
        low['5370'], # 즐거움
        low['5361'], # 열정
        low['5363'], # 행복
        low['5364'], # 슬픔
        low['5365'], # 분노
        low['5366'], # 외로움
        low['5367'], # 그리움
        low['5368'], # 두려움
        low['5369']  # 놀람
    ]
    top_emotion_data.append(top_emotion_values)
    emotion_percent_data.append(emotion_percent_values)
    
    # 1000개의 데이터가 모일때 commit
    if i % 1000 == 0:
        sql_query = """UPDATE RENEW_music_analytics 
        SET lyrics_emotion1 = %s ,lyrics_emotion2 = %s , lyrics_emotion3 = %s
        WHERE music_idx = %s"""
        cursor.executemany(sql_query,top_emotion_data)
        indj_db.commit()
        top_emotion_data = []
        
        sql_query2 = """INSERT INTO RENEW_lyrics_emotion_percent
        (music_idx , lov , joy , pas , hap , sad ,ang , lon , mis , fea , sup)
        VALUES (%s , %s , %s , %s , %s , %s , %s, %s , %s, %s, %s)"""
        cursor.executemany(sql_query2,emotion_percent_data)
        indj_db.commit()
        emotion_percent_data = []
    
# 나머지 데이터 update ( n 의 배수가 아닌 데이터)
sql_query = """UPDATE RENEW_music_analytics 
    SET lyrics_emotion1 = %s ,lyrics_emotion2 = %s , lyrics_emotion3 = %s
    WHERE music_idx = %s"""
cursor.executemany(sql_query,top_emotion_data)
indj_db.commit()

sql_query2 = """INSERT INTO RENEW_lyrics_emotion_percent
    (music_idx , lov , joy , pas , hap , sad ,ang , lon , mis , fea , sup)
    VALUES (%s , %s , %s , %s , %s , %s , %s, %s , %s, %s, %s)"""
cursor.executemany(sql_query2,emotion_percent_data)
indj_db.commit()

#연결 해제
indj_db.close()
    
```
